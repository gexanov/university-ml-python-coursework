```python
import cv2
import numpy as np

img = cv2.imread('image.jpg')
# Загрузка изображения
# Метод imread загружает изображение в формате BGR (не RGB!)

rows, cols = img.shape[:2]
# Получаем размеры изображения (количество строк и столбцов)
# img.shape возвращает кортеж (height, width, channels), нам нужны только height и width

# Поворот на 45 градусов

rotation_matrix = cv2.getRotationMatrix2D((cols/2, rows/2), 45, 1)
rotated_img = cv2.warpAffine(img, rotation_matrix, (cols, rows))
# Создаем матрицу поворота для вращения изображения на 45 градусов
# Аргументы функции getRotationMatrix2D():
# ((cols/2, rows/2)) - точка вращения (центр изображения)
# 45 - угол поворота в градусах (положительное значение - поворот против часовой стрелки)
# 1 - коэффициент масштабирования (1 означает сохранение оригинального размера)
# Применяем аффинное преобразование (вращение) к изображению
# warpAffine выполняет поворот изображения с использованием полученной матрицы поворота
# Третий аргумент (cols, rows) задает размер выходного изображения
rotated_img = cv2.warpAffine(img, rotation_matrix, (cols, rows))


# Рисуем фигуры
cv2.circle(rotated_img, (100, 100), 50, (0, 0, 255), -1)  # Красный круг
```

```python
cv2.rectangle(rotated_img, (200, 200), (300, 300), (0, 255, 0), 3)  # Зеленый прямоугольник
```

```python
import cv2
import numpy as np

img = cv2.imread("house.jpg")
h, w = img.shape[:2]


hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
hsv[:, :, 1] = hsv[:, :, 1] * 0.5
hsv[:, :, 2] = hsv[:, :, 2] * 0.9
faded = cv2.cvtColor(hsv, cv2.COLOR_HSV2BGR)

#Виньетка
kernel_x = cv2.getGaussianKernel(w, w/3)
kernel_y = cv2.getGaussianKernel(h, h/3)
mask = kernel_y * kernel_x.T
mask = mask / mask.max()
vignette = np.copy(faded)

for i in range(3):
    vignette[:, :, i] = vignette[:, :, i] * mask

# Зерно
noise = np.random.normal(0, 10, vignette.shape).astype(np.uint8)
grain = cv2.add(vignette, noise)

#  Царапины
for i in range(10):
    x1 = np.random.randint(0, w)
    x2 = np.random.randint(0, w)
    cv2.line(grain, (x1, 0), (x2, h), (255,255,255), 1)

#  Осветление центра
center_mask = mask
for i in range(3):
    grain[:, :, i] = grain[:, :, i] + (center_mask*30).astype(np.uint8)

# Сохранение и показ
cv2.imwrite("old_photo.jpg", grain)

cv2.imshow("Original", img)
cv2.imshow("Old Photo", grain)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
