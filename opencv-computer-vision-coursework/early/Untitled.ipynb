```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 50)
y = x**2

plt.plot(x, y, label='Квадратичная зависимость', color='teal', linestyle='--')

plt.legend()
plt.show()
```

```python
data = np.random.randn(5000)
plt.hist(data, bins = 50)
plt.show()
```

```python
sizes = [40, 30, 20, 10]
labels = ['футбол', 'баскетбол', 'хоккей', 'волейбол']
plt.pie(sizes, labels=labels)
plt.show()
```

```python
import seaborn as sns
import pandas as pd
data = {'Влажность': [10, 20, 50, 80]}
DataFrame = pd.DataFrame(data, index = ['нск', 'спб', 'нижний новгород', 'Сочи'])
sns.heatmap(DataFrame)
plt.show()
```

```python
x = np.linspace(-10, 10, 50)
y = np.linspace(-10, 10, 50)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
surface = ax.plot_surface(X, Y, Z)
plt.show()
```
